## ðŸ”¹ Returning Multiple Values in Python

In Python, a function can return multiple values, but **technically it returns a single tuple**.
You can then **unpack** that tuple into separate variables.

---

### Example: Function returns sum and multiplication

```python
def calculate(a, b):
    s = a + b
    m = a * b
    return s, m   # returns a tuple (s, m)

# unpacking the return values
sum_result, mul_result = calculate(5, 3)

print("Sum:", sum_result)
print("Multiplication:", mul_result)
```

**Output:**

```
Sum: 8
Multiplication: 15
```

ðŸ‘‰ Inside `calculate()`, the function returned **one object**:

```python
(8, 15)   # a tuple
```

Then in the main, it was **unpacked** into `sum_result` and `mul_result`.

---

### ðŸ”¹ Tuple unpacking

```python
x, y = (10, 20)
print(x)  # 10
print(y)  # 20
```

So when a function returns multiple values, Python is just using **tuple unpacking**.

---

## ðŸ”¹ Comparison with C++

In **C++**, a function cannot directly return multiple values unless you:

* Use `pair` / `tuple` from STL, or
* Use reference parameters (`&`), or
* Return a struct/class.

Python makes it simpler because it **implicitly packs** values into a tuple and allows unpacking.

## ðŸ”¹ 1. Unpacking a Tuple

If you have a tuple, you can assign its elements directly into variables.

```python
data = (10, 20, 30)

a, b, c = data
print(a, b, c)   # 10 20 30
```

ðŸ‘‰ The number of variables must match the number of values (unless you use `*`).

---

## ðŸ”¹ 2. Unpacking a List

It works the same way with lists:

```python
nums = [1, 2, 3]

x, y, z = nums
print(x, y, z)   # 1 2 3
```

---

## ðŸ”¹ 3. Function Return Unpacking

As we discussed:

```python
def calculate(a, b):
    return a + b, a * b   # returns a tuple

s, m = calculate(5, 3)    # unpack into two variables
print("Sum:", s, "| Multiplication:", m)
```

**Output:**

```
Sum: 8 | Multiplication: 15
```

---

## ðŸ”¹ 4. Using `*` (Extended Unpacking)

If you donâ€™t know the exact number of items, you can use `*` to collect the rest.

```python
data = (1, 2, 3, 4, 5)

a, b, *rest = data
print(a)     # 1
print(b)     # 2
print(rest)  # [3, 4, 5]
```

Or:

```python
*start, last = data
print(start)  # [1, 2, 3, 4]
print(last)   # 5
```

---

## ðŸ”¹ 5. Ignoring Values with `_`

If you donâ€™t need some values, you can unpack into `_`.

```python
data = (10, 20, 30)

x, _, z = data
print(x, z)   # 10 30
```
